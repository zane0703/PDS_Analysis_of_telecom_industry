```python
#all the impor needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#change the thean of the plot so as to make it look nicer then the default
plt.style.use('ggplot')
#declue all the function needed
def setPltSetting():
    plt.xticks(size=10)
    plt.yticks(size=10)
    plt.margins(x=0)

def showInfoAndSet(df):
    df.index.name="Date"
    #replace the unerscore with space and capitalize each word  so as to make looks nicer
    df.columns= df.columns.str.replace("_"," ").str.title()
    display(df)
    print("\n*** Shape ***\n",df.shape)
    print("\n*** index ***\n",df.index)
    print("\n*** column ***\n",df.columns)
    print("\n*** amount of non na***\n",df.count())
    print("\n*** info ****")
    df.info()
    display(df.describe())
```

<p>First i import mobile phone subscriptions from <a href="https://data.gov.sg/dataset/total-mobile-phone-subscriptions" >here</a> this data come with 2 csv file one is total and one is by type</p>

```python
df_mobile = pd.read_csv('mobile-phone-subscriptions-by-subscription-type.csv',index_col=0, parse_dates=True,na_values=0)
showInfoAndSet(df_mobile)
```

<p>As shown above, in this data frame have 3 columns 2 of which tell us the key of the data and the last is to tell us the amount of subscriptions for this dataframe I going to split then into 2 dataframe using pivot table one for the type of plan and another technology </p>

```python
df_mobile_total= pd.read_csv("total-mobile-phone-subscriptions.csv",index_col=0,parse_dates=True)
showInfoAndSet(df_mobile_total)
```

<p>In this data frame there only 1 columns which tell you the total mobile subscriptions for that day</p>
<p>Next, I import residential wired broadband from <a href="https://data.gov.sg/dataset/residential-wired-broadband-subscriptions-by-type-monthly">here</a></p>

```python
df_broadband = pd.read_csv("residential-wired-broadband-subscriptions-by-type-monthly.csv",index_col=0,parse_dates=True)
showInfoAndSet(df_broadband)
```

<p>In this data frame there 2 columns with one tell you the type of broadband and another tell the amount of subscriptions i going to use pivot table to place the number of subscriptions into  different columns by broadband connections </p>
<p>Next I import the fixed line subscriptions from <a href="https://data.gov.sg/dataset/fixed-line-subscriptions" >here</a></p>

```python
df_fixed_line = pd.read_csv("fixed-line-subscriptions.csv",index_col=0,parse_dates=True)
showInfoAndSet(df_fixed_line)
```

<p>In this data frame there only 1 columns that is the total number of subscriptions, in this dataset it inclued all phone line the start with digit "6" which mean it also inclued VOIP line that come with the internet subscriptions VOIP line that start with digit "3" will not be inclued</p> 

```python
df_fixed_line_type = pd.read_csv("fixed-line-subscriptions-by-type.csv",index_col=0,parse_dates=True)
showInfoAndSet(df_fixed_line_type)
```

<p>This dataframe is the same as the previous but split into residentia and Corporate line. I will also going to use pivot table to split for this data frame</p>
<p>Last but not less import the amount of resident-households from <a href="https://data.gov.sg/dataset/resident-households-by-tenancy-annual" >here </a></p>

```python
df_households = pd.read_csv("resident-households.csv",index_col=0,parse_dates=True)
showInfoAndSet(df_households)
```

<p>In this data frame there 2 one tell us what is the value about since all of them are resident households it doses not have used to itt</p>
<p>First, let observe what popular mobile technology</p>

```python
#drop the ana so it will not mixed with the data
df_mobile = df_mobile.dropna()
#as the values are stored in the same column i need to use pivot table to split then into diffent columns 
df_mobile_tech = df_mobile.pivot_table(values="Number Of Subscriptions",columns="Network Access Technology",index="Date",aggfunc=np.sum)
df_mobile_tech.groupby(df_mobile_tech.index.year).mean().plot.bar(figsize=(20,5))
# set y axis labels
plt.ylabel("Number of Subscriptions",size=20)
# set x axis labels
plt.xlabel("Year",size=20)
#set title
plt.title("Mobile technology Subscriber overtime",size=20)
setPltSetting()
plt.show()
```

<p>From the data above it show  in 2017 which is the latest, 4G is the most used, 2011 saw the higher for 3G even higher then 4G in 2017 and since 2G technologic had stop server somewhere in 2017 the user drop to almost 0</p>
<p>Next, let observe  what plan is currently the most popular</p>

```python
#using privot table to get the tyoe of plan
df_mobile_plan = df_mobile.pivot_table(values="Number Of Subscriptions",columns="Type Of Plan",index="Date",aggfunc=np.sum)
#group the data fram by year so as not overcrowd the bar graph
df_mobile_plan.groupby(df_mobile_tech.index.year).mean().plot.bar(figsize=(20,5))
# set y axis labels
plt.ylabel("Number of Subscriptions",size=20)
#set title
plt.title("Mobile plan being use overtime",size=20)
# set x axis labels
plt.xlabel("Year",size=20)
setPltSetting()
#show the graph
plt.show()
```

<p>From the latest date, it shows that postpaid is most popular but that not always the case because from 2007 to 2012 prepaid is more popular, there are also pre and postpaid was because there is not data of the plan used for  3G technologic when it was first stated </p>
<p>Next, let observe  the dataframe in boxplot to get more details</p>

```python
#make subplot
fig ,ax = plt.subplots(1,2,figsize=(20,5))
#plt the data intp box plot
sns.boxplot(x="Network Access Technology",y="Number Of Subscriptions",data=df_mobile,ax=ax[0],notch=True,showmeans=True, meanprops={"markerfacecolor":"green", "markeredgecolor":"green"})
sns.boxplot(x="Type Of Plan",y="Number Of Subscriptions",data=df_mobile,ax=ax[1],notch=True,showmeans=True, meanprops={"markerfacecolor":"green", "markeredgecolor":"green"})
fig.suptitle("Mobile subscriptions",size=20)
#set the size of the label
ax[0].xaxis.label.set_size(20)
ax[0].yaxis.label.set_size(20)
ax[1].xaxis.label.set_size(20)
ax[1].yaxis.label.set_size(20)
#output the data
plt.show()
```

<p>From the data above 3G technology have higher median the was due it being the in server longer than 4G and from the lower quartile it shows there is the period where 3G saw a shape jump of drop in user but 4G technology still has the higher record as there more mobile user today than last time<p>
<p>both pre paid and post paid are almost the same median with post paid having the highter racord and pre paid having highter IQR</p>
<p>Next, let observe the dataframe in violinplot to get even more details</p>


```python
#make subplot
fig ,ax = plt.subplots(1,2,figsize=(20,5))
#plt the data intp violin plot
sns.violinplot (x="Network Access Technology",y="Number Of Subscriptions",data=df_mobile,ax=ax[0])
sns.violinplot(x="Type Of Plan",y="Number Of Subscriptions",data=df_mobile,ax=ax[1])
fig.suptitle("Mobile subscriptions",size=20)
#set the size of the label
ax[0].xaxis.label.set_size(20)
ax[0].yaxis.label.set_size(20)
ax[1].xaxis.label.set_size(20)
ax[1].yaxis.label.set_size(20)
#output the data
plt.show()
```

```python
#make subplot
fig,ax = plt.subplots(figsize=(10,10))
#plot the data into pie chart
ax.pie([df_mobile_total.iloc[-1],df_fixed_line.iloc[-1]],labels=["Mobile","Fixed line"])
ax.set_title("Compare between fixed line and mobile subscriber ",size=20)
plt.show()
```

<p>Next, let see how fixed-line subscriptions have dropped throughout the year</p>

```python
#set plot size
plt.figure(figsize=(20,5))
#plot the data
plt.plot(df_fixed_line)
setPltSetting()
# set y axis label
plt.ylabel("Number of Subscriptions",size=20)
#set tite
plt.title("Number of fixed line user",size=20)
# set x axis label
plt.xlabel("Year",size=20)
plt.show()
```

<p>Hmm maybe it is still being widely used by the corporation so I decided to compare corporate and resident  subscriptions  to see if this is true</p>

```python
#as the value is all at the same column i user pivot_table to split it up
df_fixed_line_type2 = df_fixed_line_type.pivot_table(values="Number Of Subscriptions",columns="Type Of Fixed Lines",index="Date",aggfunc=np.sum)
#merge total fixed line with fixed line type then rename the column to Total
df_fixed_line_type2 = pd.merge(df_fixed_line_type2,df_fixed_line,how="left",on="Date").rename(columns={'Number Of Subscriptions': 'Total'})
#set plot size
plt.figure(figsize=(20,10))
plt.plot(df_fixed_line_type2)
setPltSetting()
#set title
plt.title("Type of fixed line users",size=20)
# set y axis label
plt.ylabel("Number of Subscriptions",size=20)
# set x axis label
plt.xlabel("Year",size=20)
#to show legend
plt.legend(df_fixed_line_type2.columns)
# output the plot
plt.show()
```

<p>Even compare between different user it shows that resident subscriptions are not going down</p>
<p>Then I realise something many telecoms have been bundling the fixed line with fibre broadband plan which may contribute to the fact so I decided to bring in the wired broadband data to see if is true</p>
<p>While we are on it let looks at the data of the fixed broadband first before I compere it with fixed line </p>
<p>First, I one to see what are amount of subscriptions of different type of broadband technologic</p>

```python
#as the value is all at the same column i user pivot_table to split it up
df_broadband2 = df_broadband.pivot_table(values="No Of Subscriptions",columns="Broadband Connections",aggfunc=np.sum,index="Date")
#plot the data
df_broadband2.plot(figsize=(20,5))
setPltSetting()
plt.legend(df_broadband2.columns)
fiber_start= df_broadband2.loc[~df_broadband2["Fibre Based"].isnull()].index[0]
#annnotate the plot 
plt.annotate("Fiber broadband start server",(fiber_start,df_broadband2.loc[fiber_start,"Fibre Based"]),xytext=(pd.to_datetime("1/1/2009"),df_broadband2.loc[fiber_start,"Fibre Based"]),arrowprops={"color":"grey"})
#set title
plt.title("Type of wire broadband",size=20)
# set y axis label
plt.ylabel("Number of Subscriptions",size=20)
# set x axis label
plt.xlabel("Year",size=20)
plt.show()
```

<p>From the data above it show that the fibre broadband only started at  around 2011 but it saw a huge increase  in subscriber whereas Digital subscriber line and cable broadband started dropping</p>
<p>next, let comper the broadband type in the histogram</p> 

```python
#make subplot
fig ,ax =plt.subplots(figsize=(20,5))
sns.distplot(df_broadband2["Cable Modem"],label="Cable Modem",ax=ax,bins=10)
sns.distplot(df_broadband2["DSL"],label="DSL",ax=ax,bins=10)
sns.distplot(df_broadband2["Fibre Based"],label="Fibre Based",ax=ax,bins=10)
#tel to show legend
ax.legend()
ax.set_title("Type of wire broadband",size=20)
# set c axis label
ax.set_xlabel("Number of Subscriptions",size=20)
plt.show()
```

<p>From the data above it show only fibre has reached above 1.2m whereas cable modem have 2nd higher at about 0.7m it also have 0.7m users most of the time</p>
<p>Next, let comper fibre broadband and fixed-line to see if is true</p>

```python
#merge broadband and fixed line so as to comper both data
df_merge = pd.merge(df_broadband2,df_fixed_line_type2["Residential Line"],on="Date",how="inner").dropna().rename(columns={'Residential Line': 'Fixed line'})
plt.figure(figsize=(20,5))
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['axes.titlesize'] = 20
sns.regplot(x="Fibre Based", y="Fixed line",data=df_merge)
plt.title("Compare between fixed and fibre broadband subscriber")
plt.show()
```

<p>From the data above it look like an increase in fibre broadband do cause fixed line subscriber but there is another factor that also helps to the fixed-line subscriber, one which is subscriber moving from cable broadband might contribute to it because you don't need to subscribe to fixed-line for cable broadband there must be a subscriber with cable broadband  that not subscribe to fixed-line and maybe the increase of resident households may contribute to it so I decided to compare it as while </p>

```python
#set plot size
fig, ax =plt.subplots(1,2,figsize=(20,5))
#plot the data into reg plot
sns.regplot(y="Cable Modem", x="Fixed line" ,data=df_merge,scatter_kws={"color": "green"}, line_kws={"color": "green"} ,ax=ax[0])
ax[0].set_title("Compare between fixed and cable broadband subscriber")
#merge househole and fixed line so as to comper both data
df_merge2 = pd.merge(df_fixed_line_type2["Residential Line"],df_households,on="Date",how="inner").rename(columns={'Residential Line': 'Fixed line',"Value":"Resident Households"})
#plot the data into reg plot
sns.regplot(y="Resident Households", x="Fixed line" ,data=df_merge2, scatter_kws={"color": "blue"},line_kws={"color": "blue"},ax=ax[1])
ax[1].set_title("Compare between fixed subscriber and resident households")
plt.show()
```

<p>From the data above it look like the drop in cable broadband and the increase in households do help to increase the number of Fixed-line subscribers</p>
